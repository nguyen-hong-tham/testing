#### Các test-case

| Test Case | Input (x) | Expected Output | Mục đích kiểm thử         |
|-----------|-----------|-----------------|----------------------------|
| TC1       | 1         | 2              | Kiểm tra nhánh (x < 10) và (x < 2) |
| TC2       | 5         | 10             | Kiểm tra nhánh (x < 10)   |
| TC3       | 10        | 20             | Kiểm tra giá trị biên của nhánh (x <10)   |

| TC5       | -3        | -6             | Kiểm tra giá trị của nhánh (x < 2)  |

Thông qua kết quả test-case, thấy rõ nhánh else if (x < 2) không bao giờ được thực hiện.
