## Mô tả bài toán và các input/output

### Mô tả bài toán

Bài toán yêu cầu giải phương trình bậc 4 dạng đặc biệt:

a·x⁴ + b·x² + c = 0

Trong đó:
- a, b, c là các số thực được nhập từ bàn phím hoặc đọc từ file.
- Đặt y = x², khi đó phương trình trở thành phương trình bậc hai:

a·y² + b·y + c = 0

Với mỗi nghiệm y ≥ 0, ta tìm được nghiệm thực của phương trình ban đầu:

x = ±√y

Chương trình sẽ xét đầy đủ các trường hợp đặc biệt và tổng quát:

#### Trường hợp đặc biệt:
- a = b = c = 0: phương trình có vô số nghiệm → return -1
- a = b = 0: phương trình vô nghiệm → return 0
- a = 0: phương trình trở thành b·y + c = 0. Nếu có nghiệm y ≥ 0 thì phương trình bậc 4 có 2 nghiệm thực → `return 2`

#### Trường hợp tổng quát (a ≠ 0):
- Tính Δ = b² - 4ac
- Nếu Δ < 0: phương trình vô nghiệm → return 0
- Nếu Δ ≥ 0: tính y₁, y₂. Với mỗi nghiệm y ≥ 0, sinh ra 2 nghiệm x = ±√y. Đếm tổng số nghiệm và trả về.


### Input (Dữ liệu vào)
- Ba số thực: a, b, c

### Output (Dữ liệu ra)
- "Infinite solutions" nếu phương trình có vô số nghiệm  
- "No solution." nếu phương trình vô nghiệm  
- "The equation has n real solution(s): x1 x2 ... xn" nếu có nghiệm thực

### Mô tả input và output của bài toán đã nhập kiểm tra :
* input: 0 0 0   => output: Infinite solutions.
* input: 0 0 5   => output: No solution.
* input: 0 2 -8  => output: The equation has 2 real solution(s): 2 -2 
* input: 1 0 1   => output: No solution.
* input: 1 -2 1  => output: The equation has 2 real solution(s): 1 -1 
* input: 1 0 -1  => output: The equation has 2 real solution(s): 1 -1 
* input: 1 -2 0  => output: The equation has 4 real solution(s): 1.41421 -1.41421 0 -0 
* input: 1 2 1   => output: No solution.




## Xây dựng các test cases kiểm tra tính đúng đắn chương trình

Để kiểm thử chương trình, ta cần xây dựng các test cases bao phủ như sau:
- Các trường hợp đặc biệt (a = 0, b = 0, c = 0). => có 3 test-case
- Trường hợp tổng quát với Δ < 0, Δ = 0, Δ > 0. => có 3 test case 
- Các giá trị biên có nghiệm âm và nghiệm dương. => có 2 test case 


=> Tổng cộng cần có ít nhất 8 test-case trong bài.

## Viết đoạn mã tự động kiểm tra chương trình cho bên dưới đúng hay sai?

```python
int main() {
    // Danh sách test cases
    vector<TestCase> tests = {
        {0,0,0,-1,"Infinite solutions."},
        {0,0,5,0,"No solution."},
        {0,1,-4,2," The equation has 2 real solution"},
        {1,0,1,0,"No solution."},
        {1,0,-1,2,"The equation has 2 real solution"},
        {1,2,1,0,"No solution."},
        {1,-2,1,2,"The equation has 2 real solution"},
        {1,-3,2,4,"The equation has 4 real solution"}

    };

    int passCount = 0;

    for (size_t i=0; i<tests.size(); i++) {
        double x[4];
        int n = solveQuartic(tests[i].a, tests[i].b, tests[i].c, x);
        cout << "Test " << i+1<<"( "<<tests[i].a<<","<< tests[i].b<<","<< tests[i].c<<")" << " (" << tests[i].description << "): ";
        if (n == tests[i].expectedSolutions) {
            cout << "PASS  (solutions = " << n << ") ";
            for (int i = 0; i < n; i++) {
                 cout << x[i] << " ";
                }
            cout<<endl;
            passCount++;
        } else {
            cout << "FAIL  (expected " << tests[i].expectedSolutions 
                 << ", got " << n << ")" << endl;
        }
    }

    cout << "\nSummary: " << passCount << "/" << tests.size() 
         << " test cases passed." << endl;

    return 0;
}
```
